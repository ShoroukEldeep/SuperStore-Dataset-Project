```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Sample - Superstore.csv", encoding='windows-1252')
```

```python
df.head()
```

```python
plt.rcParams['figure.figsize']=(10,5)
sns.countplot(x=df.Segment)
plt.show()
```

- In the Home Office Segment we observe that the count is less So convergent strategy and improvement is a necessity for this Segment

```python
plt.rcParams['figure.figsize']=(20,5)
plt.rcParams.update({'font.size':12})
sns.countplot(x='Sub-Category',data=df)
plt.show()
```

- itâ€™s very much clear that the Copiers and Machines Subcategory needs improvement

```python
plt.rcParams['figure.figsize']=(20,5)
plt.rcParams.update({'font.size':12})
sns.countplot(x='Region',data=df)
plt.show()
```

- we can conclude that the south region needs more improvement compared to others

```python
plt.rcParams['figure.figsize']=(20,5)
plt.rcParams.update({'font.size':12})
plt.title('Sales Per category')
sns.barplot(x='Category', y='Sales',data = df)
```

- In the Office supplies category we observe that the sales is less So convergent strategy and improvement is a necessity for this category 

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['ship month'] = df['Ship Date'].dt.month
sns.lineplot('ship month','Sales',data = df)
```

```python
sns.lineplot(df['Ship Date'].dt.year,df['Sales'])
```

```python
sns.lineplot(df['Order Date'].dt.month,df['Sales'])
```

```python
sns.lineplot(df['Order Date'].dt.year,df['Sales'])
```

```python
sns.barplot(x='Ship Mode',y = "Sales",data =df,palette = 'mako')
plt.show()
```

- Here we observe that the sales are high if the ship mode is standard class, and sales are low if the ship mode is either second class or same day

```python
loss_df= df[df['Profit'] < 0]
loss_df.head()
```

```python
plt.bar(loss_df['Sub-Category'],loss_df['Sales'])
plt.rcParams.update({'font.size':10})
plt.xlabel('Sub_Category')
plt.ylabel('Sales')
```

- we observed that the Sales for Fasteners, Appliances, Furnishings, and Accessories is very low

```python
plt.rcParams['figure.figsize']=(28,8)
plt.bar(df['Sub-Category'],df['Sales'])
plt.rcParams.update({'font.size':14})
plt.xlabel('Sub_Category')
plt.ylabel('Sales')
```

- When it comes to comparison in overall supermarket data, Fasteners, Labels, Furnishings, Art, paper, Envelopes sub-categories have very fewer sales thatâ€™s why it needs to be improved
