### Creating Loss Dataframe
- Create a new Dataframe where profit is negative which means loss and concentrate on these areas to improve

```python
loss_df=df[df['Profit'] < 0]
loss_df
```

```python
loss_df.describe()
```

```python
Total_loss =np.negative(loss_df['Profit'].sum())
Total_loss
```

```python
loss_df.groupby(df['Segment']).sum()
```

```python
loss_df.groupby(df['Sub-Category']).sum()
```

```python
loss_df['Sub-Category'].value_counts()
```

```python
loss_df.groupby(df['City']).sum().sort_values('Profit',ascending=True).head(10)
```

- list shows the names of the top 10 cities where loss is high

```python
loss_df.sort_values(['Sales'],ascending=True).groupby(df['Category']).mean()
```

- we observed that more loss is in the technology category

```python
df.groupby(['State']).sum()['Sales'].nsmallest(10)
```

- These are the last 10 states where sales are very less

```python
df.sort_values(['Segment'],ascending=True).groupby(df['Segment']).sum()
```

- Here Consumer segment sales might be less when compared to other segments but this is the only segment that provides the highest profits

```python
df.groupby(df['Region']).sum()
```

- Here we can see that sales are less in the South Region So to get better profits or more sales we should focus on this area too
